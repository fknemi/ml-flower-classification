```python
import tensorflow as tf
import keras
from sklearn import preprocessing, model_selection
import matplotlib.pyplot as plt
import numpy as np
```

```python
BATCH_SIZE = 32
SEED = 42
val_dataset = keras.utils.image_dataset_from_directory("flowers",
    color_mode="rgb",
    batch_size=BATCH_SIZE,
    image_size=(224, 224),
    shuffle=False,
    validation_split=0.2,
    subset="validation",
    seed=SEED)
train_dataset = keras.utils.image_dataset_from_directory("flowers",
    color_mode="rgb",
    batch_size=BATCH_SIZE,
    image_size=(224, 224),
    shuffle=True,
    validation_split=0.2,
    subset="training",
    seed=SEED)
```
